# Determine common types of documents based on titles and create new ledger with category tages

```python
import nltk
from nltk.corpus import PlaintextCorpusReader
corpus_root = 'text-data/Cambridge MacLehose FineReader OCR'
hakluyt = PlaintextCorpusReader(corpus_root, '.*')
from nltk.text import TextCollection
hakluyt_col = TextCollection(hakluyt)
import matplotlib.pyplot as plt
```

```python
# 01_16_NNE-_1246_Libellus_historicus_Johannis_de_Plano_Carpini_pp.055-094.txt

def HK_date(fileid):
    #extracts date component from Hakluyt text file name
    return(fileid[11:15])
def HK_geog(fileid):
    #extracts geography component from Hakluyt text file name
    return(fileid[6:10])
def HK_title(fileid):
    #extracts title component from Hakluyt text file name
    return(fileid[16:-15])
```

several next cells exploring common terms in titles, manually filtered and grouped for final tagging

```python
titles_text = [HK_title(title).split('_') for title in hakluyt.fileids()]

titles_starts = [title[:5] for title in titles_text]

titles_starts = [" ".join(title) for title in titles_starts]

#print(titles_starts)

titles_starts_text = " ".join(titles_starts)

#print(titles_starts_text)

titles_list = [" ".join(title) for title in titles_text]

titles_text = " ".join(titles_list)

print(titles_text.split())
```

```python
from nltk.corpus import stopwords
from nltk.probability import FreqDist
 
stop_words = set(stopwords.words('english'))

#print(stop_words)
 
filtered_titles_starts_text = []

for w in titles_starts_text.split():
#    print(w.lower(), 'in stop words is', w.lower() in stop_words)
    if w.lower() not in stop_words:
        filtered_titles_starts_text.append(w.lower())

#print(filtered_titles_starts_text)

fdist = FreqDist(filtered_titles_starts_text)

common_terms = [(term, count) for term, count in fdist.most_common(300)]
print(common_terms)
```

```python
filtered_titles_text = []

for w in titles_text.split():
#    print(w.lower(), 'in stop words is', w.lower() in stop_words)
    if w.lower() not in stop_words:
        filtered_titles_text.append(w.lower())

#print(filtered_titles_starts_text)

fdist2 = FreqDist(filtered_titles_text)

common_terms2 = [(term, count) for term, count in fdist2.most_common(300)]
print([(term, count) for term, count in common_terms2])
```

Tagging in two columns: one exclusive narrative / document column based on Hakluyt Handbook categorizing, the other based on common title terms 

```python
tags_main = ['nar', 'doc']
nar_flags = ['voyage', 'conquest', 'navigation', 'mariage', 'journal', 'landing', 'vanquishing', 'victories', 'fight', 'quicksilver', 'Carak', 'Ravillon', 'voiage', 'Angeles', 'travels', 'expedition']
tag2_groups = [
['letter', 'epistle'],
['england', 'english', 'london', 'britaine'], 
['russia', 'mosco'],
['granted', 'graunted'],
['jerusalem', 'holy_land', 'palaestina'],
['spain', 'spanish'],
['commandement', 'order', 'instruction'],
['syria', 'alepo']]
tag2_singles = ['voyage', 'discovery', 'description', 'indies', 'relation', 'testimon', 'note', 'discourse', 'guinea', 'brasil', 'persia', 'constantinople', 'prussia', 'charter', 'instructions', 'china', 'turke', 'virginia', 'privileges', 'commission', 'ambassage', 'egypt', 'report', 'patents', 'florida', 'norway', 'alger', 'barbary', 'ruttier', 'east_ind','west_ind', 'portug', 'tripolis', 'fight','canada', 'espanna', 'guiana', 'article', 'observation', 'purpose', 'asia', 'pasport','alexandria', 'marocco','newfoundland', 'history', 'conduct', 'catalogue', 'oration', 'agreement']
```

```python
import csv
with open ('text-data/ledger.csv', newline='') as ledgercsv:
    ledger = csv.reader(ledgercsv)
    with open ('text-data/ledgertaggedtest.csv', 'w', newline='') as ledgertaggedcsv:
        ledgertagged = csv.writer(ledgertaggedcsv)
        ledgertagged.writerow(['vol', 'chap', 'geog', 'date', 'title', 'pages', 'tag1', 'tag2'])
        for row in ledger:
            #skip header
            if not row == ['vol', 'chap', 'geog', 'date', 'title', 'pages']:
                row_w = row
                #determine tag1
                for tag1flag in nar_flags:
                    if tag1flag in row[4].lower():
                        row_w.append('nar')
                        break #if narrative tag spotted, tag as narrative and end tag1 loop
                # if no narrative tag spotted, tag as document
                if not row_w[-1] == 'nar':
                    row_w.append('doc')
                #determine tag2
                tag2 = ""
                # go through tag2groups
                for tags in tag2_groups:
                    for tag in tags: #in each group of tags
                        if tag in row[4].lower(): #if any member of the group found in title
                            tag2 += (", " + tags[0]) #tag as the first representative of group
                            break #and don't check the rest
                # go through tag2singles
                for tag in tag2_singles: #each single tag found in title gets added to tag2
                    if tag in row[4].lower():
                        tag2 += (", " + tag)
                #write tag2
                #remove leading comma
                if tag2: #if not empty
                    tag2 = tag2[2:]
                row.append(tag2)
                ledgertagged.writerow(row)
```
